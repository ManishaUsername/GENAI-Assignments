1: Import Libraries

```python
import random
from collections import defaultdict
```

2: Training Text

```python
text = """
Artificial intelligence is the future of technology.
Machine learning is part of artificial intelligence.
Deep learning is a powerful technique in machine learning.
Python is widely used in artificial intelligence.
"""
```

3: Preprocessing Text

```python
words = text.lower().split()
print(words)   # Just to see processed words
```

4: Build Markov Chain Model

```python
markov_chain = defaultdict(list)

for i in range(len(words) - 1):
    current_word = words[i]
    next_word = words[i + 1]
    markov_chain[current_word].append(next_word)

print(markov_chain)   # To see the chain
```

5: Text Generator Function

```python
def generate_text(start_word, length=20):
    if start_word not in markov_chain:
        return "Start word not found in training data."

    result = [start_word]
    current = start_word

    for _ in range(length - 1):
        next_words = markov_chain[current]
        if not next_words:
            break
        current = random.choice(next_words)
        result.append(current)

    return " ".join(result)
```

6: Run the Generator

```python
start = input("Enter starting word: ")
print("\nGenerated Text:")
print(generate_text(start, 30))
```
